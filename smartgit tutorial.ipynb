```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define a dataset
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
# Step 3: Visualize Data
# For simplicity, we'll visualize the ages in a bar chart
plt.figure(figsize=(10, 6))  # Setting the figure size
plt.bar(df['Name'], df['Age'], color='skyblue')  # Creating a bar chart
plt.xlabel('Name')  # Label for X-axis
plt.ylabel('Age')  # Label for Y-axis
plt.title('Age of Individuals')  # Title of the chart
plt.show()  # Display the chart
```
