```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Create another simple DataFrame
# Define a new dataset
data = {
    'Month': ['January', 'February', 'March', 'April'],
    'Sales': [200, 240, 310, 400],
    'Expenses': [150, 200, 220, 300]
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)

```

```python
# Visualize Data
# We'll visualize the Sales and Expenses over the months in a line plot for comparison
plt.figure(figsize=(10, 6))  # Setting the figure size
plt.plot(df['Month'], df['Sales'], marker='o', label='Sales')  # Line plot for Sales
plt.plot(df['Month'], df['Expenses'], marker='s', label='Expenses')  # Line plot for Expenses
plt.xlabel('Month')  # Label for X-axis
plt.ylabel('Amount')  # Label for Y-axis
plt.title('Monthly Sales and Expenses')  # Title of the chart
plt.legend()  # Add legend to distinguish the lines
plt.grid(True)  # Add grid for better readability
plt.show()  # Display the chart
```
